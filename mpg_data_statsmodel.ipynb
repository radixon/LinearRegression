```python
# Select the 'int64' columns
mask = mpg_data.dtypes == np.int64
categorical_cols = mpg_data.columns[mask]
```

```python
# Determine how many extra columns would be created
num_ohc_cols = (mpg_data[categorical_cols]
                .apply(lambda x: x.nunique())
                .sort_values(ascending=False))


# No need to encode if there is only one value
small_num_ohc_cols = num_ohc_cols.loc[num_ohc_cols>1]

# Number of one-hot columns is one less than the number of categories
small_num_ohc_cols -= 1

# The number of columns, assuming the original ones are dropped. 
small_num_ohc_cols.sum()
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Copy of the data
mpg_data_ohc = mpg_data.copy()

# The encoders
le = LabelEncoder()
ohc = OneHotEncoder()

for col in num_ohc_cols.index:
    
    # Encode the int64 categories
    dat = le.fit_transform(mpg_data_ohc[col]).astype(np.int64)
    
    # Remove the original column from the dataframe
    mpg_data_ohc = mpg_data_ohc.drop(col, axis=1)

    # One hot encode the data--this returns a sparse array
    new_dat = ohc.fit_transform(dat.reshape(-1,1))

    # Create unique column names
    n_cols = new_dat.shape[1]
    col_names = ['_'.join([col, str(x)]) for x in range(n_cols)]

    # Create the new dataframe
    new_df = pd.DataFrame(new_dat.toarray(), 
                          index=mpg_data_ohc.index, 
                          columns=col_names)

    # Append the new data to the dataframe
    mpg_data_ohc = pd.concat([mpg_data_ohc, new_df], axis=1)
```

```python
from sklearn.model_selection import train_test_split

y_col = 'mpg'

# Split the data that is not one-hot encoded
feature_cols = [x for x in mpg_data.columns if x not in ["mpg","origin","model_year","cylinders","car_name"]]
X_data = mpg_data[feature_cols]
y_data = mpg_data[y_col]

X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, 
                                                    test_size=0.3, random_state=42)

# Split the data that is one-hot encoded
feature_cols = [x for x in mpg_data_ohc.columns if x not in ["mpg","car_name"]]
X_data_ohc = mpg_data_ohc[feature_cols]
y_data_ohc = mpg_data_ohc[y_col]

X_train_ohc, X_test_ohc, y_train_ohc, y_test_ohc = train_test_split(X_data_ohc, y_data_ohc, 
                                                    test_size=0.3, random_state=42)
```
